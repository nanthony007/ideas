```python
import pandas as pd
import os
import seaborn as sns

sns.set_theme()
```

```python
os.listdir('fncs_data')
```

```python
def generate_file_names() -> list[str]:
    file_names = []
    for f in os.listdir('fncs_data'):
        file_names.append(os.path.join("fncs_data", f))
    return file_names


def combine_files() -> pd.DataFrame:
    """Combines all of the individual files into one composite file."""
    large_df = pd.concat((pd.read_csv(f) for f in generate_file_names()))
    large_df.to_csv("composite.csv", index=False)
    return large_df
```

```python
df = combine_files()
```
